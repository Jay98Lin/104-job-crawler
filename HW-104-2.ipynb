```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = "https://www.104.com.tw/jobs/search/?keyword=資料工程師&jobsource=2018indexpoc&ro=0"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

jobs = []

# 遍歷每個職缺
for idx, job in enumerate(soup.find_all('article', class_='job-list-item')):
    # 如果是前兩列熱門推薦工作職缺，跳過
    if idx < 2:
        continue
    
    # 提取工作抬頭
    job_title = job.find('a', {'data-qa-id': 'jobSeachResultTitle'}).text.strip()
    
    # 提取公司名稱
    company_name = job.find('a', title=re.compile('公司名：')).text.strip()

    # 提取工作地點、經驗要求、學歷要求
    job_info = job.find('ul', class_='job-list-intro').find_all('li')
    location = job_info[0].text.strip() if len(job_info) > 0 else ''
    experience = job_info[1].text.strip() if len(job_info) > 1 else ''
    education = job_info[2].text.strip() if len(job_info) > 2 else ''

    # 提取薪水
    salary_info = job.find('div', class_='job-list-tag').find('span', class_='b-tag--default')
    salary = salary_info.text.strip() if salary_info else '薪水資訊未提供'

    # 提取網址
    job_link = 'https://www.104.com.tw' + job.find('a', {'data-qa-id': 'jobSeachResultTitle'})['href']

    jobs.append([job_title, company_name, location, experience, education, salary, job_link])

df = pd.DataFrame(jobs, columns=['工作標題', '公司名稱', '工作地點', '經驗要求', '學歷要求', '薪水', '網址'])
df
```
